```python
# 假设 datas 变量已经通过读取 "KuaiRand_train.txt" 文件加载完毕
datas = []
file_path = "Food_train.txt" # 将文件名存储在一个变量中，方便复用

# 1. 从文件中读取原始数据
try:
    with open(file_path, 'r') as f:
        for line in f:
            datas.append(line.strip().split(' '))
except FileNotFoundError:
    print(f"错误: 文件 '{file_path}' 未找到。请确保文件存在于正确的路径。")
    # 如果文件不存在，则退出程序
    exit()

augmented_data = []

# 2. 对序列进行前缀增强
for item_sequence in datas:
    if len(item_sequence) < 2:
        continue

    for i in range(1, len(item_sequence)):
        prefix = item_sequence[:i+1]
        augmented_data.append(prefix)

# 3. 打印统计信息（可选）
print(f"原始数据样本数量: {len(datas)}")
print(f"增强后数据样本数量: {len(augmented_data)}")
if augmented_data:
    print("\n增强过程完成，准备将数据写入文件...")
else:
    print("\n没有生成可用于增强的数据。")


# 4. 将增强后的数据写回到原始文件
try:
    with open(file_path, 'w') as f:
        for seq in augmented_data:
            if len(seq) >= 2:
                # 将列表中的每个 item_id（通常是字符串）用空格连接起来
                line_to_write = ' '.join(seq)
                # 写入文件，并在末尾添加换行符
                f.write(line_to_write + '\n')
    print(f"\n成功！增强后的 {len(augmented_data)} 条序列已保存到 '{file_path}'。")

except IOError as e:
    print(f"\n错误：无法写入文件 '{file_path}'。错误信息: {e}")
```
