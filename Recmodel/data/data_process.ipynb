```python
file_name = "gowalla.txt"
Max_item = 0
user_num = 0
sum_inter = 0
with open(file_name) as f:
    for line in f:
        user_num += 1
        seq = line.strip().split(' ')[1:]
        sum_inter += len(seq)
        for i in range(len(seq)):
            seq[i] = int(seq[i])
            Max_item = max(Max_item, seq[i])
avg = sum_inter / user_num
density = sum_inter / (user_num * Max_item)
sparse = 1 - density
print("user_num: ", user_num)
print("Max_item: ", Max_item)
print("sum_inter: ", sum_inter)
print("avg: ", avg)
print("density: ", density)
print("sparse: ", sparse)
        
```
