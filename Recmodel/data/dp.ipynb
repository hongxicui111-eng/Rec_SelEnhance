```python
import os

# 获取当前目录下的所有txt文件
txt_files = [f for f in os.listdir() if f.endswith('train.txt')]

# 打开并读取所有txt文件的内容
for file_name in txt_files:
    datas = []
    with open(file_name, 'r', encoding='utf-8') as file:
        data = [line.strip().split(' ') for line in file]
        for dat in data:
            temp = []
            for item in dat:
                temp.append(item)
                if len(temp) >= 2: datas.append(str(temp))
                # print(datas[-1]==temp)
    with open(file_name, 'w', encoding='utf-8') as file:
        for line in datas:
            # 重新组合每行的内容并写入文件
            file.write(' '.join(eval(line)) + '\n')
datas[1]
```

```python
import random

# 读取数据
with open('tmall-buy.txt', 'r') as f:
    datas = [line.strip().split(' ')[1:] for line in f]

# 打乱数据
random.shuffle(datas)
datas = datas[:20000]
# 计算各个数据集的大小
train_size = int(len(datas) * 0.7)
val_size = int(len(datas) * 0.2)
test_size = len(datas) - train_size - val_size

# 划分数据集
train_data = datas[:train_size]
val_data = datas[train_size:train_size+val_size]
test_data = datas[train_size+val_size:]

# 写入到文件
with open("Tmall_train.txt", 'w') as f_train:
    for line in train_data:
        f_train.write(' '.join(line) + '\n')

with open("Tmall_val.txt", 'w') as f_val:
    for line in val_data:
        f_val.write(' '.join(line) + '\n')

with open("Tmall_test.txt", 'w') as f_test:
    for line in test_data:
        f_test.write(' '.join(line) + '\n')

print("数据已按 7:2:1 的比例划分并保存。")
```
